```python
import numpy as np
import math

def round_half_up(n, decimals = 2):
    
    multiplier = 10 ** decimals
    
    return math.floor( n * multiplier + 0.5 ) / multiplier

def round_up(n, decimals = 2): 
    
    rounded_abs = round_half_up( abs(n), decimals )
    
    return math.copysign( rounded_abs, n )
```

```python
# completely inside: 
# x_min <= x <= x_max AND y_min <= y <= y_max
# completely outside
# partially inside
```

```python
# find position of line's end-points: (x1,y1) and (x2,y2)
# region code: 
# bit_3 = above, bit_2 = below, bit_1 = right, bit_0 = left

def calculate_outcode(x, y, x_min, x_max, y_min, y_max):
    
    bit_0, bit_1, bit_2, bit_3 = 0, 0, 0, 0
    
    if x < x_min:
        
        bit_0 = 1
        
    if x > x_max:
        
        bit_1 = 1
    
    if y < y_min:
        
        bit_2 = 1
    
    if y > y_max:
        
        bit_3 = 1
    
    return str(bit_3) + str(bit_2) + str(bit_1) + str(bit_0)
```

```python
def find_intersection_point_of_line(x1, y1, x_min, x_max, y_min, y_max, m):
    
    outcode = calculate_outcode(x1, y1, x_min, x_max, y_min, y_max)
    # bit_3, bit_2, bit_1, bit_0
    above, below, right, left = outcode[:]
    # as string character, not integer
    
    x, y = 0, 0
    
    #if ABOVE boundary intersection
    if int(above) == 1:
        
        print("applying above intersection")
        
        y = y_max
        print("y:", format(y, ".2f"))
        
        x = x1 + ( 1/m * (y_max - y1) )
        print("x:", format(x, ".2f"))
    
    
    #if BELLOW boundary intersection
    elif int(below) == 1:
        
        print("applying below intersection")
        
        y = y_min
        print("y:", format(y, ".2f"))
        
        x = x1 + ( 1/m * (y_min - y1) )
        print("x:", format(x, ".2f"))
    
    
    #if RIGHT boundary intersection
    elif int(right) == 1:
        
        print("applying right intersection")
        
        x = x_max
        print("x:", format(x, ".2f"))
        
        y = y1 + ( m * (x_max - x1) )
        print("y:", format(y, ".2f"))
    
    
    #if LEFT boundary intersection
    elif int(left) == 1:
        
        print("applying left intersection")
        
        x = x_min
        print("x:", format(x, ".2f"))
        
        y = y1 + ( m * (x_min - x1) )
        print("y:", format(y, ".2f"))
    
    return x,y
```

```python
x_min, y_min, x_max, y_max = -205, -73, 266, 117

x1, y1 = -200, -2

x2, y2 = 349, -137

outcode_1 = calculate_outcode(x1, y1, x_min, x_max, y_min, y_max)
print(outcode_1)

outcode_2 = calculate_outcode(x2, y2, x_min, x_max, y_min, y_max)
print(outcode_2)

```

```python
while(True):
    print("outcode_1:", outcode_1)
    print("outcode_2:", outcode_2)
    
    a1, b1, r1, l1 = outcode_1
    a2, b2, r2, l2 = outcode_2
    oc = str(int(a1) * int(a2)) + str(int(b1) * int(b2)) + str(int(r1) * int(r2)) + str(int(l1) * int(l2))
    print("outcode_1 AND outcode_2 ==== oc:", oc)
    
    if (outcode_1 == '0000') and (outcode_2 == '0000'):
        print("Completely Inside")
        print("completely inside with coordinates x1= {}, y1= {}, x2= {}, y2= {}".format(x1, y1, x2, y2))
        break
        
    elif oc != '0000':
        print("Completely Outside")
        print("completely Outside with coordinates x1= {}, y1= {}, x2= {}, y2= {}".format(x1, y1, x2, y2))
        break
        
    else:
        print("partially inside")
        
        m = round_up((y2 - y1) / (x2 - x1))
        print("m", format(m, ".2f"))
        
        if outcode_1 != "0000":
            print("outcode_1 != '0000'")
            
            x1, y1 = find_intersection_point_of_line(x1, y1, x_min, x_max, y_min, y_max, m)
            print("new x1:", x1, "new y1:", y1)
            
            outcode_1 = calculate_outcode(x1, y1, x_min, x_max, y_min, y_max)
            print("outcode_1: re-calculated ", outcode_1)
            
        else:
            print("outcode_2 != '0000'")
            
            x2, y2 = find_intersection_point_of_line(x2, y2, x_min, x_max, y_min, y_max, m)
            print("new x2:", x2, "new y2:", y2)
            
            outcode_2 = calculate_outcode(x2, y2, x_min, x_max, y_min, y_max)
            print("outcode_2: re-calculated", outcode_2)
        
        continue
```
