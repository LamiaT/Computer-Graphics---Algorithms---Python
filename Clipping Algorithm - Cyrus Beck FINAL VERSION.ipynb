```python
# clipping window
x_min, y_min = ( -264,  -127)
x_max, y_max = ( 266,  123)

# points
x0, y0 = -318, 174
x1, y1 = 320, -161

print("(x0, y0) = ({}, {})".format(x0, y0))
print("(x1, y1) = ({}, {})".format(x1, y1))

D = (x1 - x0, y1 - y0)
print("D: ", D)
```

```python
import numpy as np

t_E = 0
t_L = 1

print("For Left Boundary::")
Ni = (-1, 0)
print("Ni:", Ni)
dot_prod_left = np.dot(Ni, D)
print("Ni.D: ", dot_prod_left)

if dot_prod_left != 0:  # ignore edges parallel to line
    
    t_left = (-(x0 - x_min)) / (x1 - x0)
    print("t_left:", round(t_left,2))

    if dot_prod_left < 0:
        print("PE or PL:  PE")
        t_E = max(t_E, t_left)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))
    
    else:
        print("PE or PL:  PL")
        t_L = min(t_L, t_left)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))


print()
print("For Right Boundary::")
Ni = (1, 0)
print("Ni:", Ni)
dot_prod_right = np.dot(Ni, D)
print("Ni.D: ", dot_prod_right)

if dot_prod_right != 0: # ignore edges parallel to line
    
    t_right = (-(x0 - x_max)) / (x1 - x0)
    print("t_right:", round(t_right,2))

    if dot_prod_right < 0:
        print("PE or PL:  PE")
        t_E = max(t_E, t_right)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))
    
    else:
        print("PE or PL:  PL")
        t_L = min(t_L, t_right)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))

print()
print("For Bottom Boundary::")
Ni = (0, -1)
print("Ni:", Ni)
dot_prod_bottom = np.dot(Ni, D)
print("Ni.D: ", dot_prod_bottom)

if dot_prod_bottom != 0: # ignore edges parallel to line
    
    t_bottom = (-(y0 - y_min)) / (y1 - y0)
    print("t_bottom:", round(t_bottom,2))

    if dot_prod_bottom < 0:
        print("PE or PL:  PE")
        t_E = max(t_E, t_bottom)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))
    
    else:
        print("PE or PL:  PL")
        t_L = min(t_L, t_bottom)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))

print()
print("For Top Boundary::")
Ni = (0, 1)
print("Ni:", Ni)
dot_prod_top = np.dot(Ni, D)
print("Ni.D: ", dot_prod_top)

if dot_prod_top != 0: # ignore edges parallel to line
    
    t_top = (-(y0 - y_max)) / (y1 - y0)
    print("t_top:", round(t_top,2))

    if dot_prod_top < 0:
        print("PE or PL:  PE")
        t_E = max(t_E, t_top)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))
    
    else:
        print("PE or PL:  PL")
        t_L = min(t_L, t_top)
        print("t_E:", round(t_E,2))
        print("t_L:", round(t_L,2))
```

```python
if t_E > t_L:
    print("Here, t_E = {} and t_L = {}".format( round(t_E,2), round(t_L,2) ))
    print("So, t_E > t_L")
    print("Line segment is OUTSIDE clip window")
    
else:
    print("Here, t_E = {} and t_L = {}".format( round(t_E,2), round(t_L,2) ))
    print("True clip intersection::")
    P_tE = (x0 + (round(t_E,2) * D[0]), y0 + (round(t_E,2) * D[1]))
    print("P_tE:", P_tE)
    P_tL = (x0 + (round(t_L,2) * D[0]), y0 + (round(t_L,2) * D[1]))
    print("P_tL:", P_tL)
    print()
    print("{} and {} are the endpoints of the clipped line.".format(P_tE, P_tL))
```
