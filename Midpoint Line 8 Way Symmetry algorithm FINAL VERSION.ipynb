```python
def find_zone(x0, y0, xn, yn):
    zone, dx, dy = 0, xn - x0, yn - y0
    print("dx = ", dx, " dy = ", dy)
    
    if abs(dx) > abs(dy): 
        if dx > 0 and dy > 0:
            zone = 0
        elif dx < 0 and dy > 0:
            zone = 3
        elif dx < 0 and dy < 0:
            zone = 4
        elif dx > 0 and dy < 0:
            zone = 7        
    else:
        if dx > 0 and dy > 0:
            zone = 1
        elif dx < 0 and dy > 0:
            zone = 2
        elif dx < 0 and dy < 0:
            zone = 5
        elif dx > 0 and dy < 0:
            zone = 6
    
    return zone
```

```python
def convert_to_zone_0(x0, y0, xn, yn, zone):
    X1, Xn, Y1, Yn = x0, xn, y0, yn # by default, zone 0
    
    if zone == 1:    # x = +Y, y = +X
        X1, Xn, Y1, Yn = y0, yn, x0, xn
    
    elif zone == 2:  # x = +Y, y = -X
        X1, Xn, Y1, Yn = y0, yn, -x0, -xn
    
    elif zone == 3:  # x = -X, y = +Y
        X1, Xn, Y1, Yn = -x0, -xn, y0, yn
    
    elif zone == 4:  # x = -X, y = -Y
        X1, Xn, Y1, Yn = -x0, -xn, -y0, -yn
    
    elif zone == 5:  # x = -Y, y = -X
        X1, Xn, Y1, Yn = -y0, -yn, -x0, -xn
    
    elif zone == 6:  # x = -Y, y = +X
        X1, Xn, Y1, Yn = -y0, -yn, x0, xn
    
    elif zone == 7:  # x = +X, y = -Y
        X1, Xn, Y1, Yn = x0, xn, -y0, -yn
    
    return X1, Y1, Xn, Yn
```

```python
def original_zone(X, Y, zone):
    
    x, y = X, Y
    
    if zone == 1:     # x' = +Y, y' = +X
        x, y = Y, X
    
    elif zone == 2:   # x' = -Y, y' = +X
        x, y = -Y, X
    
    elif zone == 3:   # x' = -X, y' = +Y
        x, y = -X, Y
    
    elif zone == 4:   # x' = -X, y' = -Y
        x, y = -X, -Y
    
    elif zone == 5:   # x' = -Y, y' = -X
        x, y = -Y, -X
    
    elif zone == 6:   # x' = +Y, y' = -X
        x, y = Y, -X
    
    elif zone == 7:   # x' = +X, y' = -Y
        x, y = X, -Y
        
    
    print("Original Zone::  x= {} , y= {} ".format(x, y))
    print()
```

```python
def mpl_zone_0(x0, y0, xn, yn, zone):
    dx, dy = xn - x0, yn - y0
    print("Zone 0:: dx = ", dx, "Zone 0:: dy = ", dy)
    
    d = (2 * dy) - dx
    print("d: ", d)
    
    increment_NE = 2 * (dy - dx)
    print("increment_NE: ", increment_NE)
    increment_E = 2 * dy
    print("increment_E", increment_E)
    
    x, y = x0, y0
    
    while x <= xn: # choosing E / NE
        print(" MLP Zone 0::  x= {} , y= {} ".format(x, y))
        original_zone(x, y, zone)
        x += 1
        
        if d <= 0: # Choose E
            d += increment_E # updating d according to E
            print("Updated d: ", d)
        
        else: # Choose NE
            y += 1
            d += increment_NE # updating d according to NE
            print("Updated d: ", d)
```

```python
def my_mpl8ws_algo(x0, y0, xn, yn):
    
    zone = find_zone(x0, y0, xn, yn)
    print("Zone: ", zone)
    
    a,b,c,d = convert_to_zone_0(x0, y0, xn, yn, zone)
    print("Converted to Zone 0 Coordinates: ")
    print("x0 = {} , y0 = {} , xn = {} , yn = {}".format(a,b,c,d))
    print()
    
    mpl_zone_0(a,b,c,d,zone)
```
