```python
import numpy as np
import math

def round_half_up(n, decimals = 2):
    
    multiplier = 10 ** decimals
    
    return math.floor( n * multiplier + 0.5 ) / multiplier

def round_up(n, decimals = 2): 
    
    rounded_abs = round_half_up( abs(n), decimals )
    
    return math.copysign( rounded_abs, n )
```

```python
def pixelize(x, y):
    print("x: ", x, " y: ", y)
```

```python
def my_dda(x0, y0, xn, yn):
    #3
    dx, dy = xn - x0, yn - y0
    
    m = round_up(dy / dx, 2)
    print("m: ", m)
    
    b = round_up(y0 - (m * x0), 2)
    print("b: ", b)
    
    #4
    number_of_steps = max(abs(dx), abs(dy))

    #5
    x, y = x0, y0
    pixelize(round_up(x,2), round_up(y,2))
    
    
    for count in range(int(number_of_steps)):
    
        if m >= -1 and  m <= 1:
            if xn >= x0:
                x += 1
                y += m
                pixelize(round_up(x,2), round_up(y,2))
                
            else:
                x -= 1
                y -= m
                pixelize(round_up(x,2), round_up(y,2))
                #   
        else:
            print("1/m: ", format(1/m, ".2f"))
            if yn >= y0:
                x += round_up(1/m, 2)
                y += 1
                pixelize(round_up(x,2), round_up(y,2))
                
            else:
                x -= round_up(1/m, 2)
                y -= 1
                pixelize(round_up(x,2), round_up(y,2))
                
```
