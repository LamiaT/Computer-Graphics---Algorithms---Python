```python
def midpoint_circle(radius, origin = (0,0)):
    
    origin_x, origin_y = origin 
    
    x = 0 
    y = radius
    print(" x = {}, y = {}".format(x + origin_x, y + origin_y))
    
    d = 1- radius
    print("d: ", d)
    
    while x < y:
        
        if d < 0:
            print("Choose E")
            
            d = d + (2 * x) + 3
            print("updated d: ", d)
            
            x += 1
            
        else:
            print("Choose SE")
            
            d = d + (2 * x) - (2 * y) + 5
            print("updated d: ", d)
            
            x += 1
            
            y -= 1
        
        print()
        print(" x = {}, y = {}".format(x + origin_x, y + origin_y))
        print("d: ", d)
```

```python
radius = 15
# origin = (0, 0)
origin = (30, 20) # NOT Centered at origin now

midpoint_circle(radius, origin)
```
